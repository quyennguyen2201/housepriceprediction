```python
from rasterio.plot import show
from rasterio.merge import merge
import rasterio as rio
from pathlib import Path
from os import listdir
from os.path import isfile, join
import os 
```

```python
path = 'C:/Users/ngquy43p/OneDrive - University of Otago/Stranded Real Asset Project/04 Data/01 SLR & Groundwater/LiDAR data/'
folder_list =[x[0] for x in os.walk(path)] 
folder_list
```

```python
for folder in folder_list[1:]: 
    input_path = folder
    output_path =  folder.split('-GTiff')[0] + '.tif'
    print(input_path)
    print(output_path)
    
    #Check if file exisst 
    if os.path.exists(output_path):
        print('Exist already')
    else:
        print('Start merging')
        raster_files =  [f for f in listdir(input_path) if  '.tif' in f]
        raster_to_mosiac = []

        for i,p in enumerate(raster_files):
            print(i/len(raster_files )* 100, '%')
            file_location = input_path +'/'+ p 
            raster = rio.open(file_location)
            raster_to_mosiac.append(raster)

        mosaic, output = merge(raster_to_mosiac)

        output_meta = raster.meta.copy()
        output_meta.update(
            {"driver": "GTiff",
                "height": mosaic.shape[1],
                "width": mosaic.shape[2],
                "transform": output,
            }
        )

        with rio.open(output_path, 'w', **output_meta) as m:
            m.write(mosaic)
```
